# While Loop

```python
age = 15
while age < 18:
    print('Student',age)
    age = age + 1
```

```python
time = 0
while time <= 10:
    print(time)
    time = time + 1
```

```python
butoon = 'hello'

while butoon != 'exit':
    print(butoon)
    butoon = input("Eeter 'exit' to stop loop")
```

# BREAK

```python
for i in range(1,6):
    if i == 3:
        break
    else:
        print(i)
```

```python
for i in 'hello':
    if i == 'l':
        break
    print(i)
```

# continue

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
    print('Hello')
```

```python
for i in 'hello':
    if i == 'l':
        continue
    print(i, 'python')
    print(i, 'python')
    print(i, 'python')
    print(i, 'python')
```

```python
roll = [1,2,3,4,5,6]
```

```python
for i in roll:
    if i == 1:
        continue
    elif i == 4:
        continue
    print(f'Homework 10 times for roll no {i}')
```
